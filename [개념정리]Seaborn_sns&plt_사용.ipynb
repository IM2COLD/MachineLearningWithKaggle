### Seaborn 사용법
- plt(matplotlib.pyplot)를 사용하여 그림의 크기, 제목, 축 레이블 및 다양한 스타일과 속성을 지정할 수 있습니다.

- sns.barplot 함수를 사용하여 막대 그래프를 그릴 때에도 내부적으로는 Matplotlib이 사용되고 있습니다.
- plt를 직접적으로 호출하지 않고, sns.barplot 함수만 사용 가능.
- sns.barplot만 사용하여 Seaborn의 기능을 활용하여 막대 그래프를 생성하고 표시할 수 있습니다.
---
- Seaborn 함수를 호출할 때마다 기본적으로 Seaborn은 새로운 Matplotlib figure와 axes를 생성합니다.
- 만약 여러 그래프를 같은 figure에 그리고 싶다면, 해당 figure와 axes를 명시적으로 관리해주어야 합니다.

> - Seaborn 함수를 호출하면 기본적으로 그래프를 그릴 새로운 figure와 axes가 생성됩니다. 그런 다음에 plt.show()를 호출하면 현재까지 생성된 figure와 axes에 있는 그래프가 표시됩니다.
- 만약 이전의 그래프를 지우고 싶다면 plt.clf()를 사용하거나, plt.close()를 사용하여 현재의 figure를 닫은 후에 새로운 그래프를 그릴 수 있습니다.



```python
import seaborn as sns
import matplotlib.pyplot as plt

# 첫 번째 그래프
sns.barplot(x=[1, 2, 3], y=[4, 5, 6])
plt.title('Bar Plot 1')
plt.show()

# 현재의 figure를 닫고 새로운 figure를 생성
plt.close()

# 두 번째 그래프
sns.lineplot(x=[1, 2, 3], y=[7, 8, 9])
plt.title('Line Plot 1')
plt.show()
```



````python
# 두 개의 서로 다른 그래프를 같은 figure에 그릴 때
import seaborn as sns
import matplotlib.pyplot as plt
# 첫 번째 그래프
sns.barplot(x=[1, 2, 3], y=[4, 5, 6])
# 두 번째 그래프
sns.lineplot(x=[1, 2, 3], y=[7, 8, 9])
# 그래프를 표시
plt.show()
````

````python
import seaborn as sns
import matplotlib.pyplot as plt
# 새로운 figure와 axes 생성
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
# 첫 번째 그래프를 첫 번째 axes에 그림
sns.barplot(x=[1, 2, 3], y=[4, 5, 6], ax=axes[0])
# 두 번째 그래프를 두 번째 axes에 그림
sns.lineplot(x=[1, 2, 3], y=[7, 8, 9], ax=axes[1])
# 그래프를 표시
plt.show()
````

- 이렇게 하면 두 개의 그래프가 같은 figure에 그려집니다. plt.subplots() 함수를 사용하여 여러 axes를 생성하고, 각각의 Seaborn 함수 호출 시에 ax 매개변수를 이용하여 특정 axes에 그리면 됩니다.

### < 서로 다른 그래프르 순차적으로 그리고 싶을 때 >
- 각각의 plt.show()를 호출하기 전에 Seaborn 함수 호출을 하면 됩니다. 각각의 함수 호출은 새로운 figure와 axes를 생성하게 됩니다.

````python
import seaborn as sns
import matplotlib.pyplot as plt

# 첫 번째 그래프
sns.barplot(x=[1, 2, 3], y=[4, 5, 6])
plt.title('Bar Plot 1')
plt.show()

# 두 번째 그래프
sns.lineplot(x=[1, 2, 3], y=[7, 8, 9])
plt.title('Line Plot 1')
plt.show()
#  각각의 plt.show() 호출 이후에 새로운 figure가 생성되며, 그래프들이 순차적으로 표시됩니다.
````
