### open 함수의 사용

- open 함수는 파일을 열 때 사용되는 내장 함수

```python
for r,_,*s,_ in[*open(0)][1:]:
  print(''.join(c*int(r)for c in s))
# [*...]는 리스트로 변환합니다.
# for r,_,*s,_ in ...: 리스트의 각 요소에 대해 반복하는데, 각 요소는 행에서 공백으로 구분된 값들입니다.
# r은 첫 번째 값, s는 나머지 값을 나타냅니다. 언더스코어 _는 해당 위치의 값을 무시합니다.
# *s는 나머지 값을 리스트로 가져옵니다.
```
- 이 코드에서 open(0)은 표준 입력을 의미합니다. open 함수의 인자로 파일 경로를 전달하면 해당 파일을 열게 되지만, 0을 전달하면 표준 입력을 의미합니다.
- 표준 입력은 사용자가 키보드로 입력한 데이터를 읽어들입니다.
- 이 코드는 표준 입력에서 데이터를 읽어온 후, 리스트로 변환하고 각 행에 대해 처리하여 결과를 출력합니다.
- 여기서 [*open(0)][1:]는 표준 입력에서 읽어들인 데이터를 리스트로 변환한 후, 첫 번째 요소를 제외한 나머지 행들을 선택합니다.

> 종합하면, 이 코드는 파일에서 데이터를 읽어와 각 행의 처리 결과를 출력합니다. 행은 첫 번째 값 r이 반복 횟수를 나타내고, 나머지 값 s는 해당 문자를 나타냅니다. 결과는 각 행에서 문자를 반복한 결과를 모두 이어붙여 출력합니다.

---
#### **open 함수 사용 예시**

```python
with open('example.txt', 'r') as file:
    for line in file:
        print(line.strip())
```
- with open(...) as ... 구문을 사용하면 파일을 열고 작업을 마치면 자동으로 파일을 닫아줍니다.
- 'r'은 읽기 모드를 나타냅니다.
- 'w'는 쓰기 모드를 의미하며, 'a'는 추가 모드(기존 내용을 유지하고 새로운 내용을 덧붙임)를 나타냅니다.

### >>> 4 : 4 비트 shift

```python
print(sum(min(5*ord(x)>>4,27)-17 for x in input()))
```
- for x in input(): 사용자로부터 문자열을 입력받아, 문자열의 각 문자에 대해 반복합니다.
- 5*ord(x) >> 4: 각 문자 x에 대해 ASCII 코드 값을 가져와서 5를 곱하고, 오른쪽으로 4 비트 시프트(shift) 합니다. ord(x) 함수는 문자 x의 ASCII 코드를 반환합니다.

-  \>> 4는 오른쪽으로 4비트 시프트 연산을 수행하며, 이는 값을 2의 4승(즉, 16)으로 나누는 것과 같습니다. 이 연산의 결과는 정수로 나오게 됩니다. 즉, 소수 부분은 버리고 정수 부분만을 취합니다.

- 만약 122 * 5를 16으로 나누는 과정에서 소수 부분이 발생하면, 시프트 연산은 소수 부분을 제거하여 정수 부분만을 취하게 됩니다. 따라서 122 * 5 \>> 4의 결과는 정수가 됩니다.

> 4비트를  shift한 이유

-
4비트를 시프트하는 부분은 비트 연산자를 사용하여 원하는 연산을 수행하는 부분입니다. 5 * ord(x) >> 4에서 >>는 오른쪽 시프트 연산자를 나타냅니다.

- 여기서 4비트 오른쪽 시프트를 하는 이유는 원래 값을 16으로 나누는 효과를 가지기 때문입니다. 2의 거듭제곱으로 나누는 것은 비트 시프트를 통해 더 효율적으로 할 수 있습니다.

- 실제로 5 * ord(x) >> 4는 다음과 같은 계산을 수행합니다:

- 5 * ord(x): 문자 x의 ASCII 코드 값에 5를 곱합니다.
- \>> 4: 이 결과를 4비트 오른쪽으로 시프트합니다. 이는 값을 2^4(즉, 16)으로 나누는 것과 동일한 효과를 갖습니다.
이렇게 4비트 오른쪽 시프트를 통해 값을 나누는 것은 특정 비트 연산을 통해 수치 계산을 빠르게 수행할 수 있도록 하는 최적화 방법 중 하나입니다.
