>  **Matplotlib에서 `plt.subplots()`, `plt.figure()`와 같은 코드를 사용하면 새로운 figure를 초기화하게 됩니다. 이는 새로운 figure와 새로운 axes 객체가 생성되어 이전에 그렸던 그래프들이 사라지게 됩니다.**

- **`plt.subplots()`**: 이 함수는 새로운 figure와 하나의 axes 객체를 생성합니다. 따라서 이전에 그렸던 그래프들은 사라지게 됩니다.

  ```python
  import matplotlib.pyplot as plt
  
  fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (6,3))  # 새로운 figure와 axes 생성
  # 여기에 새로운 그래프를 그리거나 다른 작업 수행
  
  plt.show()
  ```

- **`plt.figure()`**: 이 함수는 새로운 figure를 생성합니다. 이 때 `figsize` 매개변수를 사용하여 figure의 크기를 지정할 수 있습니다. 이 함수를 사용하면 현재의 figure를 초기화하고 새로운 figure가 정의됩니다.

  ```python
  import matplotlib.pyplot as plt
  
  figure = plt.figure(figsize=(5, 2))  # 새로운 크기를 가진 figure 생성
  # 여기에 새로운 그래프를 그리거나 다른 작업 수행
  
  plt.show()
  ```

따라서 위와 같은 코드들을 사용하면 이전에 그렸던 그래프들이 사라지고 새로운 figure가 정의됩니다.

#### **< Matplotlib에서 새로운 figure와 axes를 생성하여 이전에 그렸던 그래프들을 초기화하는 예시 코드 >**

1. **`plt.clf()`**: 이 함수는 현재 figure의 모든 그래프 요소를 지웁니다. 새로운 그래프를 그리기 전에 이 함수를 호출하여 현재 figure를 초기화할 수 있습니다.

   ```python
   import matplotlib.pyplot as plt
   
   # 이전 그래프 그리기
   
   plt.clf()  # 현재 figure 초기화
   
   # 새로운 그래프 그리기
   ```

2. **`plt.cla()`**: 이 함수는 현재 axes의 모든 그래프 요소를 지웁니다. 이전 그래프를 그리는데 사용된 axes를 초기화할 수 있습니다.

   ```python
   import matplotlib.pyplot as plt
   
   # 이전 그래프 그리기
   
   plt.cla()  # 현재 axes 초기화
   
   # 새로운 그래프 그리기
   ```

3. **`plt.close()`**: 이 함수는 현재 figure를 닫습니다. 이는 모든 그래프 요소를 제거하고 figure를 삭제합니다.

   ```python
   import matplotlib.pyplot as plt
   
   # 이전 그래프 그리기
   
   plt.close()  # 현재 figure 닫기
   
   # 새로운 그래프 그리기
   ```

이러한 함수들을 사용하여 이전에 그렸던 그래프를 제거하고 새로운 그래프를 그릴 수 있습니다.


### **Matplotlib Pyplot 모듈의 이해**

````python
# 간단한 직선 그래프

import matplotlib.pyplot as plt
plt.plot([1,2,3],[2,4,6]) # 내부적으로 Axes.plot()을 호출해서 그래프를 그림
plt.title("Hello plot") # 마찬가지로 Axes.set_title()을 호출
plt.show()
````

> - 코드에서 figure()나 subplots()를 직접 호출하지 않아도 됩니다.
- Matplotlib은 기본적으로 현재 활성화된 figure와 그 안에 포함된 subplot을 사용하여 그림을 그립니다. (**Matplotlib은 현재 활성화된 figure에 대한 기본 axes를 사용하여 그래프를 그립니다.**)
- 만약 figure나 subplot을 명시적으로 정의하지 않으면, Matplotlib은 자동으로 필요한 요소들을 생성하며, **만약 figure나 subplot을 명시적으로 정의하지 않았다면, Matplotlib은 기본적으로 첫 번째 subplot을 사용합니다.**
- 위의 코드에서는 plt.plot() 함수가 호출될 때, Matplotlib은 현재 활성화된 figure와 그 안에 있는 subplot을 사용하여 그래프를 그립니다.
- 그러나 만약 여러 개의 서로 다른 그래프를 그릴 때나 더 복잡한 레이아웃을 원할 때는 figure()나 subplots()를 사용하여 figure와 subplot을 명시적으로 정의하는 것이 유용할 수 있습니다.

> ### **Figure와 Axes**

- **Figure : 그림을 그리기 위한 캔버스(도화지)의 역할** => 그림판의 사이즈를 조정한다든가, 플롯을 최종적으로 화면에 나타내는 역할 but, 그림을 그리는 역할은 하지 않아, 그림을 그리는 실제 작업과는 연관이 없다.
- **Axes : 실제 그림을 그리는 메서드들을 가짐**, 대부분의 시각화 수행 메서드들은 axes에서 호출됨, x,y축 타이틀, 범례 등의 속성을 설정 => **이름으로 인해 단순히 축과 관련된 설정으로 오해 x**

````python
# plt.figure()는 주로 figure의 크기를 조절하는 데 사용됨.
# 아래는 figure 크기가 가로 5, 세로 2인 Figure객체를 설정하고 반환함.
plt.figure(figsize = (5,2))

## 기본으로 설정된 Axes에서 Axes.plot()함수를 호출하여 그림을 그린다. Axes.plot()은 다시 line2D 객체를 호출하여 선 그래프를 그린다. ##
plt.plot([1,2,3], [4,5,6])

## plt.title() : 내부적으로 다시 Axes.set_title() 함수를 호출하여 타이틀을 설정함 ##
plt.title("Hello plot2")
plt.show()
````

- 따라서, 변수에 할당하지 않고 단독으로 사용하더라도 현재 활성화된 figure를 정의하는데 문제는 없습니다.

- pyplot이 import되어 있는 상태에서 plt.axes()를 호출하면, 주피터 노트북은 plt.show()가 없어도 자동으로 plt.show()와 같은 역할을 하는 함수를 호출하면서 그림이 표출되게함

> - plt.axes() 함수를 호출하면 새로운 subplot을 생성합니다. 이 subplot은 현재 활성화된 figure에 추가됩니다. 만약 figure가 없으면 새로운 figure가 생성되고 그 안에 subplot이 추가됩니다. **또한, pyplot 모듈이 import 되어 있는 상태에서 plt.axes()함수를 실행시키면, 주피터 노트북은 별도의 plt.show()가 없더라도 axes가 할당되어 있으면 자동으로 plt.show()와 같은 역할을 하는 함수를 호출하기에 그림이 표출된다.**
- 주피터 노트북에서는 "inline" 백엔드가 기본으로 설정되어 있습니다. 이는 그림이 생성될 때 자동으로 출력되는 설정입니다. 따라서 plt.show() 함수를 호출하지 않아도 코드 셀의 실행 결과에 그림이 나타납니다. => **plt.axes() 함수를 사용하여 subplot을 생성하면 해당 subplot이 현재 활성화된 figure에 추가됩니다. 이때 주피터 노트북의 "inline" 백엔드가 자동으로 생성된 그림을 표시하므로 별도의 plt.show() 호출 없이 그림이 출력됩니다.**

- **이때 반환된 ax 변수는 tuple이 아니라 넘파이 ndarray 형태이며, 각각 개별 원소인 ax[0]는 왼쪽의 axes 객체를, ax[1]은 오른쪽의 axes객체를 가리킨다.**

> 여러 개의 subplots들을 이용해서 시각화 구현 시 유의할 점 : **개별 Axes 객체를 직접 이용해서 시각화를 구현해야 한다**

````python
import numpy as np

x_value_01 = np.arange(1, 10)
x_value_02 = np.arange(1, 20)
y_value_01 = 2 * x_value_01
y_value_02 = 2 * x_value_02

# figsize는 (6,3)인 Figure와 2개의 Axes를 가지는 subplots를 반환
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(6, 3))

# 개별 Axes객체들을 각각 이용하여 선 그래프와 막대그래프를 그림.
ax[0].plot(x_value_01, y_value_01, color='red', marker='o', linestyle='dashed', label='red line')
ax[1].bar(x_value_02, y_value_02, color='green', label='bar plot')

# 개별 Axes 객체 각각에 X축명을 설정
ax[0].set_xlabel('ax[0] x axis')
ax[1].set_xlabel('ax[1] x axis')

# 개별 Axes객체 각각에 범례를 설정
ax[0].legend()
ax[1].legend()

# 개별 Axes객체 각각에 타이틀을 설정
ax[0].set_title('Hello line')
ax[1].set_title('Hello bar')

plt.show()
````

> subplots(nrows = 2, ncols = 2, figsize = (5,5)) 와 같이 nrows, ncols 모두 2이상인 경우, 1차원 배열 인덱싱 방식이 아닌 2차원 배열 인덱싱 방식으로 plot.subplots()로 반환된 Axes 객체에 접근해야 한다.

````python
import numpy as np

x_value_01 = np.arange(1, 10)
x_value_02 = np.arange(1, 20)
y_value_01 = 2 * x_value_01
y_value_02 = 2 * x_value_02

# figsize는 (6,6)인 Figure와 2x2개의 Axes를 가지는 subplots을 반환.
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (6,6))

# 2x2 Axes객체들을 각각 이용하여 선 그래프와 막대그래프를 그림.
ax[0][0].plot(x_value_01, y_value_01, color='red', marker='o', linestyle='dashed', label='red line')
ax[0][1].bar(x_value_02, y_value_02, color='green', label='green bar')
ax[1][0].plot(x_value_01, y_value_01, color='green', marker='o', linestyle='dashed', label='green line')
ax[1][1].bar(x_value_02, y_value_02, color='red', label='red bar')

# 개별 Axes객체 각각에 X축명을 설정.
ax[0][0].set_xlabel('ax[0][0] x axis')
ax[0][1].set_xlabel('ax[0][1] x axis')
ax[1][0].set_xlabel('ax[1][0] x axis')
ax[1][1].set_xlabel('ax[1][1] x axis')

# 개별 Axes객체 각각에 범례를 설정
ax[0][0].legend()
ax[0][1].legend()
ax[1][0].legend()
ax[1][1].legend()

plt.show()
````
