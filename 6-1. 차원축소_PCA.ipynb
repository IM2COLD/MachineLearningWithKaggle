> Feature 선택 vs Feature 추출

- 선택 : 종속성이 강한 불필요한 feature는 제거하고 data의 특징을 잘 나타내는 주요 피처만을 선택
- 추출 : 기존 feature를 저차원의 중요 피쳐로 압축해서 추출하는 것 -> 기존의 feature가 압축된 것이므로 기존의 feature와는 완전히 다른 값
- 단순 압축이 아닌 피쳐를 함축적으로 더 잘 설명할 수 있는 또 다른 공간으로 mapping해 추출하는 것 => **기존 feature가 전혀 인지하기 어려웠던 잠재적인 요소 Latent Factor를 추출하는 것을 의미**


> ### 차원 축소 Dimension Reduction : 주어진 데이터의 특징은 캡처, 특성은 줄이기

- 차원 공간에서 데이터 포인트들 간의 거리는 측정하기 어려워지고, 공간 내의 대부분의 영역이 데이터로 채워지지 않게 되어 희소한(Spase) 구조를 가지게 됩니다. 이는 차원이 증가함에 따라 데이터 포인트들이 서로 더 멀리 떨어지는 경향이 있기 때문에 발생합니다.

- 예를 들어, 2차원 평면에서는 두 점 간의 거리를 직선 거리로 간단하게 계산할 수 있습니다. 그러나 고차원 공간에서는 두 점 간의 거리를 계산하기 위해 많은 계산이 필요하며, 두 점 간의 거리가 더 멀어질 가능성이 높아집니다. 또한, 많은 차원에서는 데이터가 특정 축에 집중되어 있지 않고 전체 공간에 균등하게 퍼져 있을 수 있으므로, 데이터 간의 상대적인 거리가 점점 더 멀어지게 됩니다.

- 이러한 현상은 고차원 데이터에서의 거리 측정, 클러스터링, 시각화, 및 모델링 등에 영향을 미칠 수 있습니다. 따라서 고차원 데이터를 다룰 때는 주의가 필요하며, 차원 축소나 특성 선택 등의 기술을 사용하여 데이터를 관리하는 것이 중요합니다.

-> 차원 축소는 데이터의 양을 줄이는 것이 아니라, 주어진 데이터의 특징(Feature)의 수를 줄이는 것을 의미합니다. 따라서 원래 데이터셋의 관측치의 수는 동일하게 유지됩니다. 정보를 보존하면서 데이터의 노이즈를 제거하거나, 더 간결하고 해석하기 쉬운 형태로 표현하기 위해 한다.

> **수백 개 이상의 feature로 구성된 데이터셋에서 상대적으로 적은 차원에서 학습된 모델에 비해 예측 신뢰도가 낮아지는(정확도가 떨어진는) 이유**
1. 차원의 저주(Curse of Dimensionality): 고차원 데이터에서는 데이터 포인트 간의 거리가 증가하고 데이터의 분포가 희소해지는 경향이 있습니다. 이로 인해 학습 데이터가 부족하게 되고, 모델이 새로운 데이터에 대해 일반화하는 데 어려움을 겪을 수 있습니다.
2. 과적합(Overfitting): 고차원 데이터에서는 모델이 학습 데이터에 과도하게 적합하여 학습 데이터에 대해 높은 정확도를 보이지만, 새로운 데이터에 대한 일반화 성능이 낮아지는 경향이 있습니다. 이는 불필요한 특성들로 인해 모델이 데이터의 잡음이나 불필요한 패턴을 학습하게 되는 것과 관련이 있습니다.
3. 고차원 데이터의 시각화와 해석의 어려움: 고차원 데이터를 시각화하거나 해석하는 것은 어려울 수 있습니다. 따라서 모델이 데이터의 패턴을 이해하고 해석하는 데에도 어려움을 겪을 수 있습니다.
4. 차원 축소의 필요성: 고차원 데이터에서는 차원 축소가 필요한 경우가 많습니다. 이는 특성 선택 또는 차원 축소 기술을 사용하여 데이터의 차원을 줄이고 모델의 복잡성을 줄이는 것을 의미합니다.

- 수백 개 이상의 feature로 구성된 data set의 경우, 상대적으로 적은 차원에서 학습된 model보다 예측 신뢰도가 떨어집니다.

- 또한, feature가 많을 경우 개별 피쳐 간에 상관관계가 높을 가능성이 큽니다. => 다중 공선성 문제

-> 따라서 수백 개 이상의 feature로 구성된 데이터셋에서는 적절한 전처리와 모델 선택이 매우 중요합니다. 모델 선택 시에는 과적합을 방지하기 위해 규제화(Regularization) 기법을 사용하고, 특성 선택이나 차원 축소 기법을 사용하여 데이터를 줄이는 것이 바람직합니다. 또한, 데이터의 특성을 잘 이해하고 적절한 특성 공학을 수행하여 모델의 성능을 향상시키는 것이 중요합니다.

-> 차원 축소로 인해 일부 정보가 손실될 수 있고, 이로 인해 모델의 성능이 저하될 수 있습니다. 특히 중요한 특징이나 패턴이 유실되면 모델의 정확도가 떨어질 수 있습니다. 또한, 차원 축소가 무작위로 이루어지는 것이 아니라 적절한 방법과 매개변수를 선택하여 이루어져야 합니다. 잘못된 차원 축소 방법이나 매개변수 설정은 모델의 성능을 떨어뜨릴 수 있습니다.

> **Feature 추출 : 단순히 Data의 압축이 핵심이 아닌, 차원 축소를 통해 좀 더 data를 잘 설명할 수 있는 잠재적인 요소를 추출**

- 매우 많은 차원을 가지고 있는 이미지 or text 에서 차원 축소를 통해 잠재적인 의미를 찾아 주는 데 이런 알고리즘이 활용되고 있음

- In Image : 매우 많은 pixel로 이뤄진 이미지 data에서 잠재된 특성을 피쳐로 도출해 함축적 형태의 이미지 변환과 압축을 수행 가능 => 이렇게 변환된 이미지는 **기존 이미지보다 훨씬 적은 차원이기에 이미지 분류 등의 분류 수행 시, 과적합 영향력이 작아져서 오히려 원본 data로 예측을 수행하는 것보다 더 높은 에측 성능을 낼 수 있다**

- In Text : 텍스트 문서의 숨겨진 의미를 추출 가능
- 문서 내 단어들의 구성에서 숨겨져 있는 Sementic, Topic을 잠재 요소로 간주하고 이를 찾아내는 SVD, NMF


## **PCA (Principal Component Analysis) 주성분 분석**

- 여러 변수 간에 존재하는 상관관계를 이용해 이를 대표하는 주성분을 추출해 차원을 축소하는 기법
- PCA로 차원을 축소할 때는 기존 data의 정보 유실이 최소화되어야 한다. -> 가장 높은 분산을 가지는 data의 축을 찾아 이 축으로 차원을 축소하는데, 이것이 pca의 주성분이 된다.

image.png

1. 분산이 가장 큰 (data의 특징을 가장 잘 설명하는) 첫 번째 벡터 축을 생성
2. 1번재 벡터 축에 직각이 되는 벡터(직교 벡터) 중 분산이 가장 큰 벡터를 두 번째 벡터 축으로 생성
3. 2번째 벡터 축과 직각이 되는 벡터를 설정하는 방식으로 축을 생성

-> 이렇게 생성된 벡터 축에 원본 data를 투영하면 벡터 축의 개수만큼의 차원으로 원본 data의 차원이 축소됨

> **입력 data의 공분산 행렬이 고유벡터와 고유값으로 분해될 수 있으며, 이렇게 분해된 고유벡터를 이용해 입력 데이터를 선형 변환하는 방식이 PCA이다.**

1. 입력 data set의 공분산 행렬을 생성
2. 공분산 행렬의 고유벡터와 고유값을 계산
3. 고유값이 가장 큰 순서대로 k개 만큼 고유벡터를 추출
4. 고유값이 가장 큰 순으로 추출된 고유벡터를 이용해 새롭게 입력 data를 변환

```python
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

iris = load_iris()
# Numpy data set을 판다스 DataFrame으로 변환
columns = ['sepal_length','sepal_width','petal_length','petal_width']
irisDF = pd.DataFrame(iris.data, columns = columns)
irisDF['target'] = iris.target
irisDF.head(3)
```

```python
#setosa는 세모, versicolor는 네모, virginica는 동그라미로 표현
markers=['^', 's', 'o']

#setosa의 target 값은 0, versicolor는 1, virginica는 2. 각 target 별로 다른 shape으로 scatter plot
for i, marker in enumerate(markers):
    x_axis_data = irisDF[irisDF['target']==i]['sepal_length']
    y_axis_data = irisDF[irisDF['target']==i]['sepal_width']
    plt.scatter(x_axis_data, y_axis_data, marker=marker,label=iris.target_names[i])

plt.legend()
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.show()
```

> PCA는 여러 속성의 값을 연산해야 하기에 속성의 스케일에 영향을 받는다 => **여러 속성을 PCA로 압축하기 전에 각 속성값을 동일한 스케일로 변환하는 것이 필요하다.**

```python
from sklearn.preprocessing import StandardScaler
# Target 값을 제외한 모든 속성 값을 StandardScaler를 이용하여 표준 정규 분포를 가지는 값들로 변환
iris_scaled = StandardScaler().fit_transform(irisDF.iloc[:,:-1])
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 2)
pca.fit(iris_scaled)
iris_pca = pca.transform(iris_scaled)
print(iris_pca.shape)
```

```python
pca_columns = pca_columns=['pca_component_1','pca_component_2']
irisDF_pca = pd.DataFrame(iris_pca, columns=pca_columns)
irisDF_pca['target']=iris.target
irisDF_pca.head(3)
```

```python
#setosa를 세모, versicolor를 네모, virginica를 동그라미로 표시
markers=['^', 's', 'o']

#pca_component_1 을 x축, pc_component_2를 y축으로 scatter plot 수행.
for i, marker in enumerate(markers):
    x_axis_data = irisDF_pca[irisDF_pca['target']==i]['pca_component_1']
    y_axis_data = irisDF_pca[irisDF_pca['target']==i]['pca_component_2']
    plt.scatter(x_axis_data, y_axis_data, marker=marker,label=iris.target_names[i])

plt.legend()
plt.xlabel('pca_component_1')
plt.ylabel('pca_component_2')
plt.show()
```

```python
print(pca.explained_variance_ratio_)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
import numpy as np

rcf = RandomForestClassifier(random_state=156)
scores = cross_val_score(rcf, iris.data, iris.target, scoring='accuracy',cv=3)
print('원본 데이터 교차 검증 개별 정확도:',scores)
print('원본 데이터 평균 정확도:', np.mean(scores))
```

```python
pca_X = irisDF_pca[['pca_component_1', 'pca_component_2']]
scores = cross_val_score(rcf, pca_X, iris.target, scoring='accuracy',cv=3)
print('원본 데이터 교차 검증 개별 정확도:',scores)
print('원본 데이터 평균 정확도:', np.mean(scores))
```

> 더 많은 feature가 있는 data로 실습

```python
import pandas as pd

df = pd.read_excel('/content/drive/MyDrive/Kaggle - 파이썬 머신러닝 완벽 가이드/kaggleData/pca_credit_card.xls', header = 1, sheet_name = 'Data').iloc[0:,1:]
print(df.shape)
df.head(3)
```

```python
df.rename(columns = {'PAY_0' : 'PAY_1', 'default payment next month' : 'default'}, inplace = True)
y_target = df['default']
X_features = df.drop('default', axis = 1)
```

> 해당 data set은 23개의 feature가 있다. 각 속성끼리 상관도가 매우 높기에 상관도를 구해보고, heatmap으로 시각화한다.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

corr = X_features.corr()
plt.figure(figsize = (14,14))
sns.heatmap(corr, annot = True, fmt = ' .1g')
```

-> 높은 상관도를 가진 속성들은 소수의 PCA만으로도 자연스럽게 이 속성들의 변동성을 수용 가능

```python
# BILL_AMT1 ~ 6의 feature들을 2개의 feature로 변환한 뒤, 개별 component의 변동성을 확인
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

cols_bill = ['BILL_AMT'+str(i) for i in range(1,7)]

scaler = StandardScaler()
df_cols_scaled = scaler.fit_transform(X_features[cols_bill])
X_features.loc[:, cols_bill] = df_cols_scaled
pca = PCA(n_components=2)
pca.fit(df_cols_scaled)
print('PCA Component별 변동성:', pca.explained_variance_ratio_)
```

```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

rcf = RandomForestClassifier(n_estimators=300, random_state=156)
scores = cross_val_score(rcf, X_features, y_target, scoring='accuracy', cv=3 )

print('CV=3 인 경우의 개별 Fold세트별 정확도:',scores)
print('평균 정확도:{0:.4f}'.format(np.mean(scores)))
```

```python
# feature 수를 6개로 변환한 data set에서의 분류 예측
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_scaled = scaler.fit_transform(X_features)

pca = PCA(n_components = 6)
df_pca = pca.fit_transform(df_scaled)
scores_pca = cross_val_score(rcf, df_pca, y_target, scoring = 'accuracy', cv = 3)

print('CV=3 인 경우의 PCA 변환된 개별 Fold세트별 정확도:',scores_pca)
print('PCA 변환 데이터 셋 평균 정확도:{0:.4f}'.format(np.mean(scores_pca)))
```
