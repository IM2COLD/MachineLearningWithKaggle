- lambda 식
- 기호 오른편 => 반환 값
- 기호 왼편 => 입력 값
> lambda 와 if else를 동시에 사용하고 싶다면, if 절의 식보다 반환 값을 먼저 기술해야 한다.



```python
# 틀린 코드
x : if x <= 15 'Child else 'Adult'
```
```python
# 올바른 코드
lambda x : 'Child' if x <= 15 else 'Adult'
```



````python
a = [x if x < 2 else 7 for x in range(5)]
print(a)
````

````python
c = 99
b = [c if c < 90 else c+1]
b
````
