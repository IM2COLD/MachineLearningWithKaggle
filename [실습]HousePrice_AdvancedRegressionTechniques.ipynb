> Numerical Column

1. 연속형 (Continuous): 실수 값을 가지며 연속적인 값을 가지는 데이터입니다. 예를 들어, 연령, 키, 무게 등이 있습니다.

2. 이산형 (Discrete): 정수 값을 가지며, 이산적인 값을 가지는 데이터입니다. 예를 들어, 카운트 데이터(전체 학생 수, 주문량 등)가 있습니다.

3. 순서형 (Ordinal): 순서 또는 등급을 나타내는 범주형 데이터입니다. 순서형 데이터는 숫자로 표현되지만 실제로는 범주를 나타내며, 그 순서에 의미가 있습니다. 예를 들어, 학업 성적 (A, B, C 등), 등급 (1등급, 2등급 등) 등이 있습니다.

4. 이진형 (Binary): 두 가지 값을 가지며, 주로 0 또는 1로 표현되는 데이터입니다. 예를 들어, 성별 (남성/여성) 등이 있습니다.

- 명목형 피쳐 Nominal column :  범주를 나타내는 데이터이며, 값들 간의 순서나 크기가 없다
- 순서형 피쳐 Ordinal feature :  범주를 나타내지만 각 범주 간에 순서 또는 등급이 있다
image.png

- 명목형 데이터의 경우에는 각 값이 범주를 나타내므로, 그냥 평균값으로 결측값을 대체하는 것은 적절하지 않습니다. 명목형 데이터는 범주 간의 차이나 순서가 없으므로, 평균값이나 중간값과 같은 대표값을 계산하는 것은 의미가 없습니다.

- 명목형 데이터에는 각 범주의 빈도를 기준으로 결측값을 대체하는 것이 일반적으로 사용됩니다. 가장 빈번하게 나타나는 값으로 결측값을 대체하는 것이 일반적입니다. 이는 데이터를 왜곡시키지 않으면서 결측값을 채울 수 있는 좋은 방법입니다.

- 10, 20, 30, 35와 같이 범주를 나타내는 값들로 구성된 명목형 데이터는 원소들의 빈도를 고려하여 결측값을 대체해야 합니다. 만약 이러한 값들이 아닌 95/4와 같은 값으로 대체하게 되면, 데이터의 특성을 왜곡시킬 수 있고, 예측 모델에 부정적인 영향을 미칠 수 있습니다.

```python
# feature 별 회귀 계수를 시각화해서 model 별로 어떤 feature의 회귀 계수로 구성되는 지 확인
def get_top_bottom_coef(model):
    # coef_ 속성을 기반으로 Series 객체를 생성. index는 컬럼명.
    coef = pd.Series(model.coef_, index=X_features.columns)

    # + 상위 10개 , - 하위 10개 coefficient 추출하여 반환.
    coef_high = coef.sort_values(ascending=False).head(10)
    coef_low = coef.sort_values(ascending=False).tail(10)
    return coef_high, coef_low

# visualization을 위한 함수
def visualize_coefficient(models):
  #3개 회귀 모델의 시각화를 위해 3개의 컬럼을 가지는 subplot 생성
  fig, axs = plt.subplots(figsize=(36,15),nrows=1, ncols=3)
  fig.tight_layout()
  # 입력인자로 받은 list객체인 models에서 차례로 model을 추출하여 회귀 계수 시각화.
  for i_num, model in enumerate(models):
      # 상위 10개, 하위 10개 회귀 계수를 구하고, 이를 판다스 concat으로 결합.
      coef_high, coef_low = get_top_bottom_coef(model)
      coef_concat = pd.concat( [coef_high , coef_low] )
      # 순차적으로 ax subplot에 barchar로 표현. 한 화면에 표현하기 위해 tick label 위치와 font 크기 조정.
      axs[i_num].set_title(model.__class__.__name__+' Coeffiecents', size=25)
      axs[i_num].tick_params(axis="y",direction="in", pad=-120)
      for label in (axs[i_num].get_xticklabels() + axs[i_num].get_yticklabels()):
        # 각 subplot의 x축과 y축의 눈금 레이블에 대한 설정을 조정
        # get_xticklabels : 현재 subplot의 x축의 눈금 레이블을 반환
          label.set_fontsize(22)
      sns.barplot(x=coef_concat.values, y=coef_concat.index , ax=axs[i_num])

# 앞 예제에서 학습한 lr_reg, ridge_reg, lasso_reg 모델의 회귀 계수 시각화.
models = [lr_reg, ridge_reg, lasso_reg]
visualize_coefficient(models)
```

```python
from sklearn.model_selection import cross_val_score

def get_avg_rmse_cv(models) :
    for model in models :
        # 분할하지 않고 전체 데이터로 cross_val_score( ) 수행. 모델별 CV RMSE값과 평균 RMSE 출력
        rmse_list = np.sqrt(-cross_val_score(model, X_features, y_target, scoring="neg_mean_squared_error", cv = 5))
        rmse_avg = np.mean(rmse_list)
        print('\n{0} CV RMSE 값 리스트: {1}'.format( model.__class__.__name__, np.round(rmse_list, 3)))
        print('{0} CV 평균 RMSE 값: {1}'.format( model.__class__.__name__, np.round(rmse_avg, 3)))

# 앞 예제에서 학습한 lr_reg, ridge_reg, lasso_reg 모델의 CV RMSE값 출력
models = [lr_reg, ridge_reg, lasso_reg]
get_avg_rmse_cv(models)
```

- 데이터가 왼쪽으로 치우쳐 있으면 음수의 값을 가지고, 오른쪽으로 치우쳐 있으면 양수의 값을 가집니다. 정규 분포를 기준으로 하면, 왜도(skewness)가 0인 경우에는 완전한 대칭을 보이며, 0보다 작으면 왼쪽으로 치우쳐 있고(음수), 0보다 크면 오른쪽으로 치우쳐 있습니다(양수).

> ## 1. feature data set의 data 분포도 확인

### 데이터의 분포가 왜곡된 것 vs 데이터 자체가 왜곡된 것

- 데이터의 분포가 왜곡된 것:

"데이터의 분포가 왜곡된 것"은 주어진 데이터의 분포가 특정한 형태나 패턴을 따르지 않고, 일반적인 형태와 다른 형태를 나타내는 것을 의미합니다. 이는 데이터가 특정한 방식으로 불균형하게 분포되어 있거나, 정상적인 분포와는 다른 형태로 나타나는 경우를 말합니다. 이는 주로 데이터 분석 및 모델링 과정에서 고려해야 하는 문제로, 데이터의 왜도(skewness), 첨도(kurtosis), 클래스 불균형 등을 확인하여 평가합니다.

- 데이터 자체가 왜곡된 것:

"데이터 자체가 왜곡된 것"은 주어진 데이터 자체가 잘못되거나 오류가 있는 경우를 의미합니다. 이는 데이터가 잘못된 방식으로 수집되었거나, 오류가 있는 경우를 말합니다. 예를 들어, 센서의 오작동으로 인한 잘못된 데이터, 실수로 입력된 잘못된 값 등이 이에 해당합니다. 이 경우에는 데이터를 재수집하거나 오류를 수정하여 데이터의 정확성을 보장해야 합니다.

### category형 Data의 불균형 vs 왜곡된 numerical feature

- 데이터의 분포가 불균형한 경우:
예를 들어, 이진 분류 문제에서 양성 클래스(Positive)의 샘플 수가 음성 클래스(Negative)의 샘플 수보다 훨씬 많은 경우입니다. 이 경우 데이터의 분포가 불균형하다고 합니다. 예를 들어, 1000개의 데이터 중 900개가 음성 클래스에 속하고, 100개만이 양성 클래스에 속한다면, 데이터의 분포가 불균형하다고 말할 수 있습니다.

- 특정 카테고리에 해당하는 인코딩된 변수의 값이 다른 카테고리에 비해 상대적으로 높은 경우: 이 경우에는 인코딩된 변수의 값의 비율 측면에서 불균형이 발생합니다. 예를 들어, 성별을 나타내는 범주형 변수를 원핫 인코딩할 때, 남성에 해당하는 변수가 더 많은 샘플에 1로 인코딩되어 있다면, 이는 남성에 해당하는 인코딩된 변수의 값이 다른 카테고리(여성)에 비해 상대적으로 높다는 것입니다. 이 경우에는 개수 측면에서는 불균형한 분포가 아니더라도 특정 카테고리에 해당하는 인코딩된 변수의 값의 비율이 불균형하게 됩니다.

- 이 두 경우를 보다 명확하게 이해하기 위해선, 데이터의 분포가 불균형한 경우와 특정 카테고리에 해당하는 인코딩된 변수의 값의 비율이 다른 카테고리에 비해 높은 경우를 서로 구분하여 생각하는 것이 중요합니다.

> skew()

- skew()를 적용하는 숫자형 피쳐에서 원핫인코딩된 카테고리형 피처는 제외해야 한다. 카테고리형 피쳐는 코드성 피처이므로 인코딩 시 당연히 왜곡될 가능성이 높다.

- skew()함수를 적용하는 dataframe은 원핫 인코딩이 적용되지 않은 dataframe이어야 한다.=> original data에서 skew()함수를 적용해서 data의 왜곡도를 측정해야지, onehotencoding 후의 dataframe에서 왜곡도가 심한 피쳐들을 골라내려고 해서는 안 된다.

- "화장실 여부가 1로 1000건, 0으로 10건이 될 수 있지만, 이는 왜곡과는 무관하다" : "화장실 여부"가 1로 1000건, 0으로 10건이 될 수 있다는 것은, 예를 들어 주어진 데이터셋에서 "화장실이 있는지 여부"를 나타내는 특성이 있다고 가정해 봅시다. 이 특성이 1이면 화장실이 있고, 0이면 화장실이 없는 것을 의미합니다. 그런데 만약 1000건의 데이터가 화장실이 있는 경우(값이 1)이고, 10건의 데이터만 화장실이 없는 경우(값이 0)라면, 이 데이터는 매우 불균형하다고 볼 수 있습니다.

- 그러나 이 불균형은 '왜곡'과는 다릅니다. **'왜곡(skewness)'은 연속형 데이터의 분포가 얼마나 비대칭인지를 측정하는 데 사용**됩니다. 예를 들어, 수입 데이터가 정규 분포가 아니라 한쪽으로 치우쳐져 있다면 왜곡이 발생한 것입니다. 하지만 "화장실 여부"와 같은 이진형 카테고리 데이터의 경우, 왜곡이라는 개념은 적용되지 않습니다. 즉, 데이터가 불균형하다고 해서 그것이 왜곡을 나타낸다고 볼 수는 없습니다.

- 따라서 '화장실 여부'와 같은 이진형 카테고리형 피처는 skew() 함수를 적용하는데 영향을 미치지 않습니다. 이러한 피처들은 원핫 인코딩으로 처리되어야 하지만, skew() 함수를 적용할 때는 이러한 카테고리형 피처들을 제외하고 숫자형 피처들에만 적용해야 합니다.

>- 카테고리 피처에 로그 함수를 적용하고 원핫 인코딩을 적용하는 것은 해당 피처의 불균형한 분포를 해결하지 않습니다. 로그 함수는 데이터의 분포를 조정하여 스케일을 줄이지만, 그것이 카테고리 피처의 불균형한 분포를 해결하지는 않습니다. 원핫 인코딩 또한 불균형한 분포를 해결하는 것이 아니라, 카테고리 피처를 모델이 이해할 수 있는 형태로 변환하는 데 사용됩니다.

> - 원핫 인코딩은 각 카테고리를 새로운 이진 피처로 변환합니다. 이를 통해 모델이 각 카테고리를 구별하여 학습할 수 있지만, 원핫 인코딩은 카테고리의 불균형한 분포를 해결하지 않습니다. 따라서 카테고리 피처의 불균형한 분포를 해결하기 위해서는 다른 방법을 사용해야 합니다.

- 카테고리 피처의 불균형한 분포를 해결하기 위한 일반적인 방법은 다음과 같습니다:

- 언더샘플링 또는 오버샘플링: 소수 클래스를 증가시키거나 다수 클래스를 감소시켜 클래스 간의 균형을 맞출 수 있습니다.
가중치 부여: 모델 학습 시 소수 클래스에 더 큰 가중치를 부여하여 불균형을 보상할 수 있습니다.
- 다른 인코딩 방법 사용: 라벨 인코딩, 타겟 인코딩 등 다른 인코딩 방법을 사용하여 카테고리 피처를 변환할 수 있습니다.
따라서 로그 함수와 원핫 인코딩은 주로 데이터 전처리 과정에서 다른 목적을 위해 사용되며, 카테고리 피처의 불균형한 분포를 해결하는 방법은 다르게 처리해야 합니다.

> categorical 피쳐에 로그 함수를 적용하고 원핫 인코딩을 적용하는 것의 장점

1. 스케일링: 로그 함수를 사용하여 카테고리 피쳐를 변환하면 데이터의 범위를 조정할 수 있습니다. 특히, 일부 카테고리 값이 다른 값보다 큰 범위를 갖는 경우에는 로그 변환을 통해 데이터를 더 좁은 범위로 스케일링할 수 있습니다.

2. 정규화: 로그 변환은 데이터의 분포를 조정하여 정규 분포에 가깝게 만들어 줄 수 있습니다. 이는 모델이 데이터를 더 잘 이해하고 학습할 수 있도록 도와줍니다.

3. 원핫 인코딩의 적용: 로그 함수를 적용한 후에 원핫 인코딩을 적용하면, 해당 카테고리 피쳐가 모델에 적합한 형태로 변환됩니다. 이를 통해 모델이 범주형 데이터를 이해하고 처리하는 데 도움이 됩니다.

4. 이상치 감소: 로그 변환은 데이터의 이상치에 대해 덜 민감하게 만들어 줄 수 있습니다. 이는 모델이 이상치를 더 잘 처리할 수 있도록 도와줍니다.

5. 해석의 향상: 로그 변환을 적용하면 데이터의 스케일이 조정되고 정규화되어 해석이 더 쉬워질 수 있습니다. 이는 모델의 결과를 이해하고 해석하는 데 도움이 됩니다.

>- 회귀 모델의 경우에는 회귀 모형의 계수(coefficient)를 통해 각 피처의 중요도를 추정할 수 있습니다. 이 계수는 각 피처의 가중치를 나타내며, 피처의 값이 종속 변수에 얼마나 영향을 주는지를 보여줍니다. 높은 절댓값을 가진 계수는 해당 피처가 종속 변수에 미치는 영향이 크다는 것을 의미합니다.

- 회귀 모델의 경우에는 회귀 계수를 통해 각 피처의 중요도를 추정할 수 있습니다. 하지만 이러한 중요도 추정은 피처 간 상호작용이나 다중공선성 등을 고려하지 않으므로 주의가 필요합니다. 종종 피처 선택이나 다른 모델 평가 기준과 함께 사용될 수 있습니다.

```python
# 모델의 중요도 상위 20개의 피처명과 그때의 중요도값을 Series로 반환.
def get_top_features(model):
    ftr_importances_values = model.feature_importances_
    ftr_importances = pd.Series(ftr_importances_values, index=X_features.columns  )
    ftr_top20 = ftr_importances.sort_values(ascending=False)[:20]
    return ftr_top20

def visualize_ftr_importances(models):
    # 2개 회귀 모델의 시각화를 위해 2개의 컬럼을 가지는 subplot 생성
    fig, axs = plt.subplots(figsize=(24,10),nrows=1, ncols=2)
    fig.tight_layout()
    # 입력인자로 받은 list객체인 models에서 차례로 model을 추출하여 피처 중요도 시각화.
    for i_num, model in enumerate(models):
        # 중요도 상위 20개의 피처명과 그때의 중요도값 추출
        ftr_top20 = get_top_features(model)
        axs[i_num].set_title(model.__class__.__name__+' Feature Importances', size=25)
        #font 크기 조정.
        for label in (axs[i_num].get_xticklabels() + axs[i_num].get_yticklabels()):
            label.set_fontsize(22)
        sns.barplot(x=ftr_top20.values, y=ftr_top20.index , ax=axs[i_num])

# 앞 예제에서 print_best_params( )가 반환한 GridSearchCV로 최적화된 모델의 피처 중요도 시각화
models = [best_xgb, best_lgbm]
visualize_ftr_importances(models)
```

> ### 회귀 모델들의 예측 결과 혼합을 통한 최종 예측

- 개별 회귀 모델의 예측 결과값을 혼합해 이를 기반으로 최종 회귀 값을 예측
- 예측 결과 혼합
- ex) A 모델 예측결과값의 40%, B 모델 예측 결과값의 60%를 더해서 최종 회귀 값으로 예측


```python
# 릿지와 라쏘 모델 혼합

def get_rmse_pred(preds):
    for key in preds.keys():
        pred_value = preds[key]
        mse = mean_squared_error(y_test , pred_value)
        rmse = np.sqrt(mse)
        print('{0} 모델의 RMSE: {1}'.format(key, rmse))

# 개별 모델의 학습
ridge_reg = Ridge(alpha=8)
ridge_reg.fit(X_train, y_train)
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(X_train, y_train)
# 개별 모델 예측
ridge_pred = ridge_reg.predict(X_test)
lasso_pred = lasso_reg.predict(X_test)

# 개별 모델 예측값 혼합으로 최종 예측값 도출
pred = 0.4 * ridge_pred + 0.6 * lasso_pred
preds = {'최종 혼합': pred,
         'Ridge': ridge_pred,
         'Lasso': lasso_pred}
#최종 혼합 모델, 개별모델의 RMSE 값 출력
get_rmse_pred(preds)
```

```python
# XGBoost 와 LightGBM을 혼합

xgb_reg = XGBRegressor(n_estimators=1000, learning_rate=0.05,
                       colsample_bytree=0.5, subsample=0.8)
lgbm_reg = LGBMRegressor(n_estimators=1000, learning_rate=0.05, num_leaves=4,
                         subsample=0.6, colsample_bytree=0.4, reg_lambda=10, n_jobs=-1)
xgb_reg.fit(X_train, y_train)
lgbm_reg.fit(X_train, y_train)
xgb_pred = xgb_reg.predict(X_test)
lgbm_pred = lgbm_reg.predict(X_test)

pred = 0.5 * xgb_pred + 0.5 * lgbm_pred
preds = {'최종 혼합': pred,
         'XGBM': xgb_pred,
         'LGBM': lgbm_pred}

get_rmse_pred(preds)
```

> ### 스태킹 모델을 통한 회귀 예측

1. 개별적인 base model
2. 개별 기반 model의 예측 data를 학습 data로 만들어서 학습하는 최종 meta model

```python
from sklearn.model_selection import KFold
from sklearn.metrics import mean_absolute_error

# 개별 기반 모델에서 최종 메타 모델이 사용할 학습 및 테스트용 데이터를 생성하기 위한 함수.
def get_stacking_base_datasets(model, X_train_n, y_train_n, X_test_n, n_folds ):
  # 지정된 n_folds값으로 KFold 생성.
  kf = KFold(n_splits = n_folds, shuffle = False)
  #추후에 메타 모델이 사용할 학습 데이터 반환을 위한 넘파이 배열 초기화
  train_fold_pred = np.zeros((X_train_n.shape[0], 1))
  test_pred = np.zeros((X_test_n.shape[0],n_folds))
  print(model.__class__.__name__ , ' model 시작 ')

  for folder_counter, (train_index, valid_index) in enumerate(kf.split(X_train_n)):
    # 입력된 학습 데이터에서 기반 모델이 학습/예측할 폴드 데이터 셋 추출
    print('\t 폴드 세트: ',folder_counter,' 시작 ')
    X_tr = X_train_n[train_index]
    y_tr = y_train_n[train_index]
    X_te = X_train_n[valid_index]
    #폴드 세트 내부에서 다시 만들어진 학습 데이터로 기반 모델의 학습 수행.
    model.fit(X_tr , y_tr)
    #폴드 세트 내부에서 다시 만들어진 검증 데이터로 기반 모델 예측 후 데이터 저장.
    train_fold_pred[valid_index, :] = model.predict(X_te).reshape(-1,1)
    #입력된 원본 테스트 데이터를 폴드 세트내 학습된 기반 모델에서 예측 후 데이터 저장.
    test_pred[:, folder_counter] = model.predict(X_test_n)

  # 폴드 세트 내에서 원본 테스트 데이터를 예측한 데이터를 평균하여 테스트 데이터로 생성
  test_pred_mean = np.mean(test_pred, axis=1).reshape(-1,1)

  #train_fold_pred는 최종 메타 모델이 사용하는 학습 데이터, test_pred_mean은 테스트 데이터
  return train_fold_pred, test_pred_mean
```

```python
# get_stacking_base_datasets( )은 넘파이 ndarray를 인자로 사용하므로 DataFrame을 넘파이로 변환.

X_train_n = X_train.values
X_test_n = X_test.values
y_train_n = y_train.values

# 각 개별 기반(Base)모델이 생성한 학습용/테스트용 데이터 반환
ridge_train, ridge_test = get_stacking_base_datasets(ridge_reg, X_train_n, y_train_n, X_test_n, 5)
lasso_train, lasso_test = get_stacking_base_datasets(lasso_reg, X_train_n, y_train_n, X_test_n, 5)
xgb_train, xgb_test = get_stacking_base_datasets(xgb_reg, X_train_n, y_train_n, X_test_n, 5)
lgbm_train, lgbm_test = get_stacking_base_datasets(lgbm_reg, X_train_n, y_train_n, X_test_n, 5)
```

```python
# 개별 모델이 반환한 학습 및 테스트용 데이터 세트를 Stacking 형태로 결합.
Stack_final_X_train = np.concatenate((ridge_train, lasso_train, xgb_train, lgbm_train), axis = 1) #  각 배열이 열 방향으로 연결
Stack_final_X_test = np.concatenate((ridge_test, lasso_test, xgb_test, lgbm_test), axis=1)

# 최종 메타 모델은 라쏘 모델을 적용.
meta_model_lasso = Lasso(alpha=0.0005)

# 기반 모델의 예측값을 기반으로 새롭게 만들어진 학습 및 테스트용 데이터로 예측하고 RMSE 측정.
meta_model_lasso.fit(Stack_final_X_train, y_train)
final = meta_model_lasso.predict(Stack_final_X_test)
mse = mean_squared_error(y_test, final)
rmse = np.sqrt(mse)
print('스태킹 회귀 모델의 최종 RMSE 값은:', rmse)
```

- 회귀 문제에서 Stacking 모델이 효과적인 경우

1. 다양한 종류의 기본 회귀 모델을 사용할 수 있는 경우, 이를 결합하여 더욱 강력한 예측 모델을 만들 수 있습니다.

2. 각 기본 모델이 서로 다른 특징을 잘 학습하고 있는 경우, Stacking은 이러한 다양성을 활용하여 더욱 정확한 예측을 할 수 있습니다.

3. 데이터가 복잡하거나 비선형적인 관계를 가질 때, Stacking은 다양한 모델을 결합하여 이러한 복잡성을 적절하게 모델링할 수 있습니다.

- 분류 문제에서 Stacking 모델이 효과적인 경우

1. 클래스의 불균형이나 클래스 사이의 복잡한 경계를 가지는 데이터셋에서, Stacking은 다양한 모델의 예측을 결합하여 더욱 정확한 분류를 할 수 있습니다.

2. 앙상블에 포함된 다양한 분류기가 각각 다른 측면에서 데이터를 잘 학습한 경우, 이러한 다양성은 Stacking에서 큰 장점이 될 수 있습니다.

3. 특정 모델이 잘 맞지 않는 outlier나 noise가 있는 데이터에 대해서도, Stacking은 이러한 문제를 완화할 수 있는 장점을 가질 수 있습니다.
